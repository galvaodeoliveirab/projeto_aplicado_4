# **Introdução**

A indústria automotiva brasileira é um dos pilares da economia nacional, representando aproximadamente 22% do PIB industrial do país e empregando diretamente mais de 100 mil trabalhadores nas montadoras, além de uma extensa cadeia produtiva que envolve fornecedores de autopeças, concessionárias e serviços de logística (ANFAVEA, 2025). Desde a instalação das primeiras montadoras na década de 1950, o setor passou por ciclos de crescimento expressivo, crises econômicas e transformações tecnológicas que moldaram sua trajetória ao longo de quase sete décadas.

Nos últimos anos, o setor automotivo brasileiro tem enfrentado desafios significativos: a crise econômica de 2015-2016, que reduziu a produção a patamares de uma década anterior; o impacto da pandemia de COVID-19, que paralisou fábricas em 2020; a escassez global de semicondutores em 2021-2022; e, mais recentemente, a crescente competição com veículos importados, especialmente de origem chinesa, que ampliaram sua participação no mercado nacional (ANFAVEA, 2024). Paralelamente, a transição para veículos eletrificados e a agenda de descarbonização impõem novos desafios de planejamento produtivo ao setor.

Neste contexto, a capacidade de realizar previsões precisas sobre produção e vendas de veículos torna-se estratégica tanto para as montadoras, que precisam planejar seus ciclos produtivos e estoques, quanto para formuladores de políticas públicas industriais. A análise de séries temporais oferece ferramentas robustas para capturar os padrões de tendência, sazonalidade e ciclos econômicos presentes nesses dados, permitindo a construção de modelos preditivos que auxiliem na tomada de decisão.

# **Motivações e justificativa**

A previsão de produção e vendas de veículos automotores é um problema de grande relevância econômica e social. O setor automotivo constitui um dos termômetros mais sensíveis da atividade econômica brasileira, sendo frequentemente utilizado como indicador antecedente da saúde da economia nacional. Variações na produção e nas vendas de veículos refletem diretamente as condições de crédito, renda disponível, confiança do consumidor e expectativas empresariais.

Do ponto de vista prático, modelos de previsão acurados para essa série temporal podem beneficiar múltiplos agentes: (i) montadoras, no planejamento de produção, gestão de estoques e dimensionamento da força de trabalho; (ii) governo, na formulação de políticas industriais e tributárias; (iii) fornecedores da cadeia automotiva, no planejamento de capacidade produtiva; e (iv) setor financeiro, na avaliação de crédito e análise de risco setorial.

A base de dados da ANFAVEA (Associação Nacional dos Fabricantes de Veículos Automotores), uma das mais longas séries temporais industriais disponíveis no Brasil, com dados mensais desde 1957, oferece uma oportunidade singular para a aplicação de técnicas de aprendizado de máquina e modelos estatísticos. A riqueza da série — que abrange períodos de hiperinflação, múltiplos planos econômicos, abertura comercial, crises internacionais e a pandemia — representa um desafio técnico relevante para modelagem preditiva, ao mesmo tempo em que fornece volume suficiente de dados para o treinamento de modelos como ARIMA e LSTM.

# **Descrição da base de dados**

A base de dados principal do projeto será obtida da **ANFAVEA — Associação Nacional dos Fabricantes de Veículos Automotores**, uma instituição de classe que reúne as empresas fabricantes de veículos automotores com produção no Brasil.

**Fonte:** ANFAVEA — Dados Estatísticos para Download.
**URL:** https://anfavea.com.br/site/edicoes-em-excel/
**Licença:** Dados públicos de acesso livre, sem necessidade de cadastro.

**Descrição dos dados:**

A ANFAVEA disponibiliza séries temporais mensais em formato Excel (.xlsm) contendo as seguintes variáveis:

| Variável | Descrição | Período | Granularidade |
|---|---|---|---|
| Produção | Unidades produzidas por segmento (automóveis, comerciais leves, caminhões, ônibus) | Jan/1957 – presente | Mensal |
| Emplacamentos (vendas) | Unidades vendidas no mercado interno (nacionais + importados) | Jan/1957 – presente | Mensal |
| Exportações | Unidades exportadas (montados + CKD) | Jan/1957 – presente | Mensal |
| Exportações em valor | Valor em US$ FOB das exportações | Jan/1985 – presente | Mensal |
| Emprego | Número de empregados no setor | Jan/1983 – presente | Mensal |
| Estoque | Unidades em estoque (nacionais + importados) | Jan/2023 – presente | Mensal |

**Volume de dados:** A série principal (produção) contém aproximadamente **828 observações mensais** (janeiro/1957 a fevereiro/2026), segmentadas em 5 categorias de veículos. Considerando todas as variáveis e segmentos, o dataset completo contém milhares de registros.

**Formato:** Arquivos Excel (.xlsm e .xlsx), que serão convertidos para DataFrame Pandas no notebook do projeto.

**Estrutura:** Dados organizados em formato tabular com colunas para ano, mês e valores por segmento de veículo. Cada planilha do arquivo corresponde a uma variável (produção, vendas, exportações etc.).
